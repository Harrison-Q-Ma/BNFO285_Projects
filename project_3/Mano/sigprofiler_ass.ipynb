```python
import pandas as pd, pathlib

raw  = pathlib.Path("/Users/manavchordia/BNFO285_Projects/project_3/Mano/mutation_matrix.csv")   
fixed = raw.with_suffix(".sbs96.txt")   
df = pd.read_csv(raw, sep=None, engine="python")

df.rename(columns={df.columns[0]: "MutationType"}, inplace=True)

df.set_index("MutationType", inplace=True)
df.index   = df.index.str.strip()
df.columns = df.columns.str.strip()

assert df.shape[0] == 96,  "expected 96 mutation contexts"
assert not df.index.duplicated().any()
assert not df.columns.duplicated().any()

df.to_csv(fixed, sep="\t") 

print("✓ wrote", fixed, "shape =", df.shape)
```
