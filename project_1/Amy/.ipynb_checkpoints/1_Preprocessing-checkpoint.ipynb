# Import

```python
import numpy as np
import pandas as pd
from pathlib import Path
```

# Load Files

```python
# Load File
DATA_PATH = Path('../../Data/')
data = pd.read_csv(Path(DATA_PATH, 'TCGA.HNSC.expression.txt'), sep = '\t').fillna("NaN")
metadata = pd.read_csv(Path(DATA_PATH, 'TCGA.HNSC.metadata.txt'), sep = '\t').fillna("NaN")
```

# Remove Normal Samples

```python
# remove normal samples
# 499 patients, 501 samples remaining
# patient TCGA-KU-A6H7 and TCGA-UF-A71A each has 2 samples
sample_type = data.sample_id.str.split('-').str[3]
sample_code = sample_type.str[:2].astype(int)
data_filtered = data[sample_code < 10]
data_filtered = data_filtered.set_index('sample_id')
data_filtered = data_filtered.drop(columns='patient_id')
```

```python
data_filtered
```

# Log2 Transform

```python
data_log = np.log2(data_filtered + 1)
data_log.head(5)
```

# PCA

## pre-processing

```python
# standardize by sample
# remove mean
# scaling to unit variance
from sklearn.preprocessing import StandardScaler
data_scaled = StandardScaler().fit_transform(data_log.T)
data_PCA = pd.DataFrame(data_scaled.T, index=data_log.index, columns=data_log.columns)
```

```python
from sklearn.decomposition import PCA
pca = PCA(
    n_components=100,
    svd_solver='auto',
    iterated_power='auto',
    random_state=None
)
pca_result = pca.fit_transform(data_PCA)
pca_df = pd.DataFrame(pca_result, index=data_PCA.index)
```

```python
pca.explained_variance_ratio_[:2]
```

```python
pca.explained_variance_ratio_[:100].sum()
```

```python
data_unfiltered = data.copy()
data_unfiltered = data_unfiltered.set_index('sample_id')
data_unfiltered = data_unfiltered.drop(columns='patient_id')
data_unfiltered_log = np.log2(data_unfiltered + 1)
data_unfiltered_scaled = StandardScaler().fit_transform(data_unfiltered_log.T)
data_unfiltered_PCA = pd.DataFrame(data_unfiltered_scaled.T, index=data_unfiltered_log.index, columns=data_unfiltered_log.columns)
pca_unfiltered_result = pca.fit_transform(data_unfiltered_PCA)
pca_unfiltered_df = pd.DataFrame(pca_unfiltered_result, index=data_unfiltered_PCA.index)
```
