```python
import pandas as pd
import numpy as np
```

```python
# Read in age and survival data
metadata = pd.read_csv("~/Desktop/TCGA.HNSC.metadata.txt", sep="\t")
features = ["patient_id", "age_at_initial_pathologic_diagnosis", "OS", "OS.time", "DSS", "DSS.time", "DFI", "DFI.time","gender"]
metadata_1 = metadata[features]
metadata_1 = metadata_1.rename(columns={"age_at_initial_pathologic_diagnosis": "age",
                                        "OS.time": "OS_time",
                                        "DSS.time": "DSS_time",
                                        "DFI.time": "DFI_time",
                                        "gender": "gender"})
```

```python
complementary_metadata = pd.read_csv("~/Desktop/most_updated_clinical.tsv", sep="\t")
complementary_features = ["cases.submitter_id", "cases.primary_site", "diagnoses.classification_of_tumor", "diagnoses.ajcc_pathologic_stage"]
metadata_2 = complementary_metadata[complementary_features]
metadata_2 = metadata_2.rename(columns={"cases.submitter_id": "patient_id", "cases.primary_site": "tumor_site", "diagnoses.classification_of_tumor": "tumor_class", "diagnoses.ajcc_pathologic_stage": "tumor_stage"})

replacement_map = {
    "Other and unspecified parts of tongue": "tongue",
    "Base of tongue": "tongue",
    "Other and unspecified parts of mouth": "mouth",
    "Floor of mouth": "mouth"
}
metadata_2["tumor_site"] = metadata_2["tumor_site"].replace(replacement_map)

replacement_map = {
    "Synchronous primary": "primary",
    "Subsequent Primary": "primary",
    "Prior primary": "primary"
}
metadata_2["tumor_class"] = metadata_2["tumor_class"].replace(replacement_map)

replacement_map = {
    "'--": "NA"
}
metadata_2["tumor_stage"] = metadata_2["tumor_stage"].replace(replacement_map)

metadata_2 = metadata_2.drop_duplicates(subset=["patient_id"])
```

```python
merged_metadata = pd.merge(metadata_1, metadata_2, on='patient_id')
merged_metadata.to_csv("~/Desktop/processed_metadata.txt", index=False, sep="\t")
merged_metadata
```
