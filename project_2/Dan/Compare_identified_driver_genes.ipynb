```python
dndscv_genes = pd.read_csv('./dndscv_output.csv')
```

```python
dndscv_sig_genes = dndscv_genes.loc[dndscv_genes['pglobal_cv'] < 1e-5]
dndscv_sig_genes['gene_name'] = dndscv_sig_genes['gene_name'].apply(lambda x: 'CDKN2A' if 'CDKN2A' in x else x)
```

```python
dndscv_sig_genes
```

```python
yesandyes = set(dndscv_sig_genes['gene_name']).intersection(set(our_genes['gene']))
```

```python
yesandyesmuts = original_mutation_calls.loc[original_mutation_calls['Hugo_Symbol'].apply(lambda x: x in yesandyes)]
```

```python
yesandyesmuts.groupby(by = 'Variant_Classification').count()['patient_id']
```

```python
cvyes_usno = set(dndscv_sig_genes['gene_name']).difference(set(our_genes['gene']))
```

```python
cvyes_usnomuts = original_mutation_calls.loc[original_mutation_calls['Hugo_Symbol'].apply(lambda x: x in cvyes_usno)]
```

```python
cvyes_usnomuts.groupby(by = 'Variant_Classification').count()['patient_id']
```

```python
usyes_cvno = set(set(our_genes['gene'])).difference(dndscv_sig_genes['gene_name'])
```
