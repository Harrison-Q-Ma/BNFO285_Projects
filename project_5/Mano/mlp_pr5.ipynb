```python
import pandas as pd

df_train = pd.read_csv('/content/drive/MyDrive/train_split_80.csv')
df_test = pd.read_csv('/content/drive/MyDrive/test_split_20.csv')

print("Train DataFrame Head:")
display(df_train.head())
print("\nTrain DataFrame Info:")
display(df_train.info())

print("\nTest DataFrame Head:")
display(df_test.head())
print("\nTest DataFrame Info:")
display(df_test.info())
```

## Data preprocessing

Handle missing values, encode categorical features, scale numerical features, and separate features (X) and target (y) for both training and testing datasets.


```python
print("Missing values in df_train:")
print(df_train.isnull().sum()[df_train.isnull().sum() > 0])

print("\nMissing values in df_test:")
print(df_test.isnull().sum()[df_test.isnull().sum() > 0])

cols_to_drop_explicitly = ['tobacco_smoking_history_indicator', 'alcohol_history_documented']
df_train = df_train.drop(columns=cols_to_drop_explicitly, errors='ignore')
df_test = df_test.drop(columns=cols_to_drop_explicitly, errors='ignore')


# numerical_cols = df_train.select_dtypes(include=['float64', 'int64']).columns
# for col in numerical_cols:
#     if df_train[col].isnull().any():
#         mean_val_train = df_train[col].mean()
#         df_train[col].fillna(mean_val_train, inplace=True)
#         df_test[col].fillna(mean_val_train, inplace=True)

# print("\nMissing values in df_train after imputation:")
# print(df_train.isnull().sum()[df_train.isnull().sum() > 0])

# print("\nMissing values in df_test after imputation:")
# print(df_test.isnull().sum()[df_test.isnull().sum() > 0])


# Identify categorical columns
categorical_cols = df_train.select_dtypes(include=['object']).columns.tolist()
print("\nCategorical columns:")
print(categorical_cols)

# Drop 'patient_id' from categorical columns as it's an identifier
if 'patient_id' in categorical_cols:
    categorical_cols.remove('patient_id')

df_train = pd.get_dummies(df_train, columns=categorical_cols, dummy_na=False)
df_test = pd.get_dummies(df_test, columns=categorical_cols, dummy_na=False)

# Align columns - crucial for consistent feature sets
train_cols = df_train.columns
test_cols = df_test.columns

# missing_in_test = set(train_cols) - set(test_cols)
# for c in missing_in_test:
#     df_test[c] = 0

# missing_in_train = set(test_cols) - set(train_cols)
# for c in missing_in_train:
#     df_train[c] = 0

df_test = df_test[train_cols]

print("\nTrain DataFrame Head after encoding and alignment:")
display(df_train.head())
print("\nTest DataFrame Head after encoding and alignment:")
display(df_test.head())


X_train = df_train.drop(['patient_id', 'three_year_status'], axis=1)
y_train = df_train['three_year_status']

X_test = df_test.drop(['patient_id', 'three_year_status'], axis=1)
y_test = df_test['three_year_status']

print("\nX_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

numerical_features = X_train.select_dtypes(include=['float64', 'int64']).columns

print("\nNumerical features for scaling:")
print(list(numerical_features))

# Apply standard scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit on training data and transform both training and testing data
X_train[numerical_features] = scaler.fit_transform(X_train[numerical_features])
X_test[numerical_features] = scaler.transform(X_test[numerical_features])

print("\nX_train Head after scaling:")
display(X_train.head())
print("\nX_test Head after scaling:")
display(X_test.head())
```

## Model Building


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import regularizers

reg = regularizers.l1_l2(l1=1e-4, l2=1e-5)

# Define the neural network architecture
model = Sequential([
    # Input layer and first hidden layer
    Dense(256, activation='relu', input_shape=(X_train.shape[1],),
          kernel_regularizer=reg),
    # Second hidden layer
    Dense(128, activation='relu',
          kernel_regularizer=reg),
    Dense(64, activation='relu',
          kernel_regularizer=reg),
    # Output layer for binary classification
    Dense(1, activation='sigmoid')
])

# Print the model summary
model.summary()
```

## Model training


```python
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=10, batch_size=32)
```

## Model evaluation

Metrics - accuracy, precision, recall, F1-score, auc-roc


```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")

y_pred_prob = model.predict(X_test)
y_pred_class = (y_pred_prob > 0.5).astype("int32")

precision = precision_score(y_test, y_pred_class)
recall = recall_score(y_test, y_pred_class)
f1 = f1_score(y_test, y_pred_class)
auroc = roc_auc_score(y_test, y_pred_prob)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"AUROC: {auroc:.4f}")
```

```python
from sklearn.metrics import confusion_matrix

tn, fp, fn, tp = confusion_matrix(y_test, y_pred_class).ravel()

# Calculate metrics
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

print(f"Sensitivity (Recall): {sensitivity:.4f}")
print(f"Specificity: {specificity:.4f}")
```
